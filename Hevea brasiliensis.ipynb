```python
#importo los paquetes
import Bio
```

```python
from Bio import Entrez #importo el Entrez
from Bio import SeqIO #importo el SeqIO
def Download_sequence (x):
    Entrez.email = "A.N.Other@example.com"
    handle=Entrez.efetch (db="nucleotide", rettype="fasta", retmode="text", id=x)
    record = SeqIO.read(handle, "fasta")
    handle.close()
    return (record.seq)
```

```python
AGRO= Download_sequence("NZ_CP033026.1")
AGRO
```

```python
virG = AGRO[102996:103758]
print(virG)
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
alignments = pairwise2.align.globalms(HEVEA, virG, 2, -1, -0.5, -0.1)
for a in alignments:
    print(format_alignment(*a))
```
